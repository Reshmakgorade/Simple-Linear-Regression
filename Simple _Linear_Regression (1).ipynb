# Fitting a linear regression model

```python
import statsmodels.formula.api as smf
```

```python
#OLS = Ordinary Least Squares
model=smf.ols("sunday~daily",data=newspaper).fit()
```

```python
# To see the best fit line(which is as close as possible to as many points as possible)
sns.regplot(x='daily',y='sunday',data=newspaper);
```

```python
#Coefficients

model.params
```

```python
#To see the model predictions

model.fittedvalues
```

# y= B0 + B1 * X1 + e

## B0 - Intercept
## B1 - Slope
## X1 - Independant data point
## e (error) - Vertical distance between the actual 'Sunady' value & the Sunday fitted value

```python
13.835630 + 391.952 *1.339715
```

```python
13.835630 + 516.981 *1.339715
```

```python
newspaper['fitted values']=model.fittedvalues
newspaper
```

```python
#For manual calculation
newspaper['predicated 1']=13.835630 + (newspaper['daily'] *1.339715)
newspaper
```

```python
#To generate the residual values
#Difference bewteen Actual & Fitted values

model.resid
```

```python
model.rsquared
```

```python
model.pvalues
```

```python
newspaper['Error']= model.resid
newspaper
```

```python
newspaper['Error_sq']=(newspaper['Error'])**2
newspaper
```

```python
Mean_Squared_Error=(newspaper['Error'])**2
newspaper
```

```python
Mean_Squared_Error=newspaper['Error_sq'].mean()
Mean_Squared_Error
```
